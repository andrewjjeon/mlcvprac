```python
# 1431. Kids With the Greatest Number of Candies

# There are n kids with candies. You are given an integer array candies, where each candies[i] represents 
# the number of candies the ith kid has, and an integer extraCandies, denoting the number of extra candies that you have.
# Return a boolean array result of length n, where result[i] is true if, after giving the ith kid all the extraCandies, 
# they will have the greatest number of candies among all the kids, or false otherwise. Note that multiple kids can have 
# the greatest number of candies.

class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        n = len(candies)
        result = []
        initial_greatest = max(candies)

        for i in range(n):
            if (candies[i] + extraCandies )>= initial_greatest:
                result.append(True)
            else:
                result.append(False)
        return result
```

```python
### 933. Number of Recent Calls ###

# Implement the RecentCounter class which initializes the counter with zero recent requests.
# - RecentCounter() initiializes the counter with 0 recent requests
# - ping(t) Adds a new request at time t, where t represents some time in milliseconds, and returns the number
#   of requests that has happened in the past 3000 milliseconds (including the new request). Specifically, 
#   return the number of requests that have happened in the inclusive range [t - 3000, t].

from collections import deque

class RecentCounter(object):

    def __init__(self):
        self.queue = deque()

    def ping(self, t):
        """
        :type t: int
        :rtype: int
        """
        self.queue.append(t)
        while self.queue[0] < t -3000:
            self.queue.popleft()
        
        return len(self.queue)


obj = RecentCounter()
print(obj.ping(1))
print(obj.ping(2))
print(obj.ping(3))
print(obj.ping(3001))
print(obj.ping(3002))
```

```python
### 206. Reverse Linked List ###

# Given the head of a singly linked list, reverse the list, and return the reversed list.
# Modified to include full implementation of a Singly LinkedList Class.

# Definition for singly-linked list.


class LinkedList():
    class ListNode():
        def __init__(self, val=0, next=None):
            self.val = val
            self.next = next

    def __init__(self, values):
        self.head = self.ListNode(values[0])
        curr = self.head

        for i in range(len(values)-1):
            prev = curr
            curr = self.ListNode(values[i+1])
            prev.next = curr

    def reverseList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        curr = head
        prev = None

        while curr:
            next = curr.next
            curr.next = prev
            prev = curr
            curr = next
        return prev

    def printList(self, head):
        curr = head
        list_vals = []
        while curr:
            list_vals.append(str(curr.val))
            next = curr.next
            curr = next
        print(" -> ".join(list_vals))


list = [1, 2, 3, 4, 5]
linkedlist = LinkedList(list)
linkedlist.printList(linkedlist.head)

reversedhead = linkedlist.reverseList(linkedlist.head)
linkedlist.printList(reversedhead)
```

```python
### 104. Maximum Depth of Binary Tree ###

# Given the root of a binary tree, return its maximum depth.
# Recursion Help
#  - Base Case: "Whats the simplest version of the problem that can be answered?"
#      A: an empty tree --> maxdepth=0
#  - Recursive Case:"Can I solve this problem using smaller versions of itself?"
#      A: yes, if we know the max depth of the left subtree and right subtree we can just add 1 to that.


# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
from collections import deque

class Solution(object):
    def maxDepth(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """
        if not root:
            return 0
        return 1 + max(self.maxDepth(root.left), self.maxDepth(root.right))
```

```python
### 872. Leaf-Similar Trees

# return true if the two trees have same leaf node sequences

# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def leafSimilar(self, root1, root2):
        """
        :type root1: Optional[TreeNode]
        :type root2: Optional[TreeNode]
        :rtype: bool
        """
        leafnodes1 = []
        leafnodes2 = []

        def dfs(root, leafnodes):
            if not root:
                return
            if not root.left and not root.right:
                leafnodes.append(root.val)

            dfs(root.left, leafnodes)
            dfs(root.right, leafnodes)
            
        dfs(root1, leafnodes1)
        dfs(root2, leafnodes2)

        return leafnodes1 == leafnodes2
```

```python
### 1926. Nearest Exit from Entrance in Maze ###  BFS

# BFS Search
# empty cells represented as '.'
# walls represented as '+'
# entrance = [r, c] and you can move up down left right
# exit is defined as empty cell that is at the border of the maze.
# return # of steps in shortest path from entrance to nearest exit or -1 if no such path exists.

from collections import deque

class Solution(object):
    def nearestExit(self, maze, entrance):
        """
        :type maze: List[List[str]]
        :type entrance: List[int]
        :rtype: int
        """
        directions = [(0, 1), (1, 0), (-1, 0), (0, -1)]
        queue = deque()
        steps = 0
        visited = set()
        start = tuple(entrance)
        tot_rows = len(maze)
        tot_cols = len(maze[0])

        queue.append((start, steps))
        visited.add(start)

        while queue:
            [(cr, cc), steps] = queue.popleft()

            for d in directions:
                nr, nc = cr + d[0], cc + d[1]
                if (0 <= nr < tot_rows) and (0 <= nc < tot_cols): # if valid cell in maze
                    if (maze[nr][nc] == '.') and ((nr, nc) not in visited): # if empty and not visited before
                        if ((nr == 0) or (nr == tot_rows - 1) or (nc == 0) or (nc == tot_cols - 1)) and ((nr, nc) != start): # if border cell space and not entrance
                            return steps + 1
                        queue.append(((nr, nc), steps+1))
                        visited.add((nr, nc))
        return -1

```

```python
### 994. Rotting Oranges ### BFS

# BFS Search
# 0 - empty cell
# 1 - fresh orange
# 2 - rotten orange

# Every minute, any fresh orange that is 4-directionally adjacent to a rotten orange becomes rotten.
# Return the minimum number of minutes that must elapse until no cell has a fresh orange. If this is impossible, return -1.

from collections import deque


def orangesRotting(grid):
    """
    :type grid: List[List[int]]
    :rtype: int
    """
    h = len(grid)
    w = len(grid[0])
    directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]

    queue = deque()
    minutes = 0

    fresh = 0
    for r in range(h):
        for c in range(w):
            if grid[r][c] == 2:
                queue.append((r, c))
            elif grid[r][c] == 1:
                fresh += 1

    while queue and fresh != 0:
        for i in range(len(queue)): # all the rotten orange locations in the current minute
            (cr, cc) = queue.popleft()

            for (dr, dc) in directions:
                (nr, nc) = (cr + dr, cc + dc)
                if (0 <= nr < h) and (0 <= nc < w): # if valid cell
                    if grid[nr][nc] == 1: # if fresh turn to rotten since neighbor rotten
                        grid[nr][nc] = 2
                        queue.append((nr, nc))
                        fresh -= 1
        minutes += 1
        
    if fresh == 0:
        return minutes
    else:
        return -1


grid = [[2,1,1],[1,1,0],[0,1,1]]
print(orangesRotting(grid))
```

```python
### 199. Binary Tree Right Side View ###

# Given root of Binary tree, assume you are standing on the right side of it. Return the values of the nodes you can see from top to bottom

from collections import deque
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def rightSideView(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: List[int]
        """
        if not root:
            return []
        
        queue = deque()
        queue.append(root)
        values = []

        while queue:
            levelsize = len(queue) # append the next level of tree to queue in current loop of for
            for i in range(levelsize):
                node = queue.popleft()
                
                if i == levelsize - 1: # if its rightmost in current level
                    values.append(node.val)

                if node.left: # add left children first so the last node in each level is the most right one
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
        return values


```

```python
### 1161. Maximum Level Sum of a Binary Tree ###

# Given the root of a binary tree, the level of its root is 1, the level of its children is 2, and so on.
# Return the smallest level x such that the sum of all the values of nodes at level x is maximal.


from collections import deque
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def maxLevelSum(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """
        queue = deque()
        levels = {}

        queue.append(root)
        level = 0

        while queue:
            level += 1
            levelsum = 0
            for i in range(len(queue)):
                node = queue.popleft()
                levelsum += node.val
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            levels[level] = levelsum
        
        max_v = -9999
        max_k = -9999
        for k, v in levels.items():
            if v > max_v:
                max_v = v
                max_k = k
            elif v == max_v:
                if k < max_k:
                    max_v = v
                    max_k = k
        return max_k
        
```

```python
### 374. Guess Number Higher or Lower ###

# I pick a number from 1 to n. You have to guess which number I picked.
# Every time you guess wrong, I will tell you whether the number I picked is higher or lower than your guess.
# You call a pre-defined API int guess(int num), which returns three possible results:

# -1: Your guess is higher than the number I picked (i.e. num > pick).
# 1: Your guess is lower than the number I picked (i.e. num < pick).
# 0: your guess is equal to the number I picked (i.e. num == pick).
# Return the number that I picked.

# The guess API is already defined for you.
# @param num, your guess
# @return -1 if num is higher than the picked number
#          1 if num is lower than the picked number
#          otherwise return 0
# def guess(num):

class Solution(object):
    def guessNumber(self, n):
        """
        :type n: int
        :rtype: int
        """
        left = 1
        right = n
        mid = (left + right) // 2

        while guess(mid) != 0:
            mid = (left + right) // 2
            if guess(mid) == -1: # if guess is higher
                left = left
                right = mid - 1 # -1 bcuz u shud exclude prev mid guess, u know its lower than prev mid
            elif guess(mid) == 1: # if guess is lower
                left = mid + 1 # +1 bcuz u shud exclude prev mid guess, u know its higher than prev mid
                right = right
        return mid
```

```python
### 2300. Successful Pairs of Spells and Potions ###

# You are given two positive integer arrays spells and potions, of length n and m respectively, 
# where spells[i] represents the strength of the ith spell and potions[j] represents the strength of the jth potion.

# You are also given an integer success. A spell and potion pair is considered successful if the product of their strengths is at least success.
# Return an integer array pairs of length n where pairs[i] is the number of potions that will form a successful pair with the ith spell.

class Solution(object):
    def successfulPairs(self, spells, potions, success):
        """
        :type spells: List[int]
        :type potions: List[int]
        :type success: int
        :rtype: List[int]
        """
        n = len(spells)
        m = len(potions)
        potions.sort()
        pairs = [0] * n

        for i in range(n): #spells
            
            left = 0
            right = m-1

            while left <= right: #answer could be anywhere between left and right inclusive
                mid = (left + right) // 2
                if (spells[i] * potions[mid]) < success: #not successful if product is less than success, shift search area to right
                    left = mid + 1
                    right = right
                    
                elif (spells[i] * potions[mid]) >= success: #successful if product is equal or greater than success, 
                # shift search area to left to get the smallest mid possible
                    left = left
                    right = mid - 1
                    
            #while loop ends when left > right, left will be pointing at the first valid index
            success_count = m - left
            pairs[i] = success_count
        
        return pairs
```

```python
### 162. Find Peak Element ###

# A peak element is an element that is strictly greater than its neighbors.
# Given a 0-indexed integer array nums, find a peak element, and return its index. If the array contains multiple peaks, return the index to any of the peaks.
# You may imagine that nums[-1] = nums[n] = -âˆž. In other words, an element is always considered to be strictly greater than a neighbor that is outside the array.
# You must write an algorithm that runs in O(log n) time.

class Solution(object):
    def findPeakElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n = len(nums)
        left = 0
        right = n-1

        while left <= right:
            mid = (left + right) // 2
            if ((mid - 1) < 0) and ((mid + 1) > (n - 1)): # Edge case of 1 element list [1]
                return mid
            elif (mid - 1) < 0: # Edge case of left neighbor being outside array
                if nums[mid + 1] < nums[mid]:
                    return mid
                else:
                    left = mid + 1
                    continue
            elif (mid + 1) > (n-1): # Edge case of right neighbor being outside array
                if nums[mid-1] < nums[mid]:
                    return mid
                else:
                    right = mid - 1
                    continue
            if nums[mid - 1] > nums[mid]: # if left neighbor greater, search left half
                right = mid - 1
            elif nums[mid + 1] > nums[mid]: # if right neighbor greater, search right half
                left = mid + 1
            else: # if neither greater you found the peak
                return mid


# MUCH SHORTER SOLUTION - INSTEAD OF EXPLICIT EDGE CASE HANDLING, JUST SUBSTITUTE LEFT AND RIGHT VALUE WITH -INFINITY
class Solution(object):
    def findPeakElement(self, nums):
        left, right = 0, len(nums) - 1

        while left <= right:
            mid = (left + right) // 2
            left_val = float('-inf') if mid == 0 else nums[mid - 1]
            right_val = float('-inf') if mid == len(nums) - 1 else nums[mid + 1]

            if nums[mid] > left_val and nums[mid] > right_val:
                return mid
            elif nums[mid] < right_val:
                left = mid + 1
            else:
                right = mid - 1
```
