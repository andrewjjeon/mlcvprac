```python
### 1768. Merge Strings Alternately ###
# You are given two strings word1 and word2. Merge the strings by adding letters in alternating order, starting with word1. 
# If a string is longer than the other, append the additional letters onto the end of the merged string. Return the merged string.

def mergeAlternately(word1, word2):
    """
    :type word1: str
    :type word2: str
    :rtype: str
    """
    merged_word = ""
    merge_length = min(len(word1), len(word2))

    for i in range(merge_length):
        merged_word += word1[i]
        merged_word += word2[i]
    
    if max(word1, word2, key=len) == word1:
        merged_word += word1[merge_length:]
    else:
        merged_word += word2[merge_length:]
    return merged_word


string1 = "abc"
string2 = "defgh"
print(mergeAlternately(string1, string2))
```

```python
### 933. Number of Recent Calls ###

# Implement the RecentCounter class which initializes the counter with zero recent requests.
# - RecentCounter() initiializes the counter with 0 recent requests
# - ping(t) Adds a new request at time t, where t represents some time in milliseconds, and returns the number
#   of requests that has happened in the past 3000 milliseconds (including the new request). Specifically, 
#   return the number of requests that have happened in the inclusive range [t - 3000, t].

from collections import deque

class RecentCounter(object):

    def __init__(self):
        self.queue = deque()

    def ping(self, t):
        """
        :type t: int
        :rtype: int
        """
        self.queue.append(t)
        while self.queue[0] < t -3000:
            self.queue.popleft()
        
        return len(self.queue)


obj = RecentCounter()
print(obj.ping(1))
print(obj.ping(2))
print(obj.ping(3))
print(obj.ping(3001))
print(obj.ping(3002))
```

```python
### 206. Reverse Linked List ###

# Given the head of a singly linked list, reverse the list, and return the reversed list.
# Modified to include full implementation of a Singly LinkedList Class.

# Definition for singly-linked list.


class LinkedList():
    class ListNode():
        def __init__(self, val=0, next=None):
            self.val = val
            self.next = next

    def __init__(self, values):
        self.head = self.ListNode(values[0])
        curr = self.head

        for i in range(len(values)-1):
            prev = curr
            curr = self.ListNode(values[i+1])
            prev.next = curr

    def reverseList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        curr = head
        prev = None

        while curr:
            next = curr.next
            curr.next = prev
            prev = curr
            curr = next
        return prev

    def printList(self, head):
        curr = head
        list_vals = []
        while curr:
            list_vals.append(str(curr.val))
            next = curr.next
            curr = next
        print(" -> ".join(list_vals))


list = [1, 2, 3, 4, 5]
linkedlist = LinkedList(list)
linkedlist.printList(linkedlist.head)

reversedhead = linkedlist.reverseList(linkedlist.head)
linkedlist.printList(reversedhead)
```

```python
### 104. Maximum Depth of Binary Tree ###

# Given the root of a binary tree, return its maximum depth.
# Recursion Help
#  - Base Case: "Whats the simplest version of the problem that can be answered?"
#      A: an empty tree --> maxdepth=0
#  - Recursive Case:"Can I solve this problem using smaller versions of itself?"
#      A: yes, if we know the max depth of the left subtree and right subtree we can just add 1 to that.


# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
from collections import deque

class Solution(object):
    def maxDepth(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """
        if not root:
            return 0
        return 1 + max(self.maxDepth(root.left), self.maxDepth(root.right))
```

```python
### 872. Leaf-Similar Trees

# return true if the two trees have same leaf node sequences

# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def leafSimilar(self, root1, root2):
        """
        :type root1: Optional[TreeNode]
        :type root2: Optional[TreeNode]
        :rtype: bool
        """
        leafnodes1 = []
        leafnodes2 = []

        def dfs(root, leafnodes):
            if not root:
                return
            if not root.left and not root.right:
                leafnodes.append(root.val)

            dfs(root.left, leafnodes)
            dfs(root.right, leafnodes)
            
        dfs(root1, leafnodes1)
        dfs(root2, leafnodes2)

        return leafnodes1 == leafnodes2
```

```python
### 1926. Nearest Exit from Entrance in Maze ###

# BFS Search
# empty cells represented as '.'
# walls represented as '+'
# entrance = [r, c] and you can move up down left right
# exit is defined as empty cell that is at the border of the maze.
# return # of steps in shortest path from entrance to nearest exit or -1 if no such path exists.

from collections import deque

class Solution(object):
    def nearestExit(self, maze, entrance):
        """
        :type maze: List[List[str]]
        :type entrance: List[int]
        :rtype: int
        """
        directions = [(0, 1), (1, 0), (-1, 0), (0, -1)]
        queue = deque()
        steps = 0
        visited = set()
        start = tuple(entrance)
        tot_rows = len(maze)
        tot_cols = len(maze[0])

        queue.append((start, steps))
        visited.add(start)

        while queue:
            [(cr, cc), steps] = queue.popleft()

            for d in directions:
                nr, nc = cr + d[0], cc + d[1]
                if (0 <= nr < tot_rows) and (0 <= nc < tot_cols): # if valid cell in maze
                    if (maze[nr][nc] == '.') and ((nr, nc) not in visited): # if empty and not visited before
                        if ((nr == 0) or (nr == tot_rows - 1) or (nc == 0) or (nc == tot_cols - 1)) and ((nr, nc) != start): # if border cell space and not entrance
                            return steps + 1
                        queue.append(((nr, nc), steps+1))
                        visited.add((nr, nc))
        return -1

```
