```python
### 1768. Merge Strings Alternately ###
# You are given two strings word1 and word2. Merge the strings by adding letters in alternating order, starting with word1. 
# If a string is longer than the other, append the additional letters onto the end of the merged string. Return the merged string.

def mergeAlternately(word1, word2):
    """
    :type word1: str
    :type word2: str
    :rtype: str
    """
    merged_word = ""
    merge_length = min(len(word1), len(word2))

    for i in range(merge_length):
        merged_word += word1[i]
        merged_word += word2[i]
    
    if max(word1, word2, key=len) == word1:
        merged_word += word1[merge_length:]
    else:
        merged_word += word2[merge_length:]
    return merged_word


string1 = "abc"
string2 = "defgh"
print(mergeAlternately(string1, string2))




### 1071. Greatest Common Divisor of Strings ###

# Given two strings str1 and str2, return the largest string x such that x divides both str1 and str2.

import math
class Solution:
    def gcdOfStrings(self, str1: str, str2: str) -> str:
        len1 = len(str1)
        len2 = len(str2)

        def isgcd(k):
            if len1 % k != 0 or len2 % k != 0:
                return False # not divisible at all
            n1 = int(len1 / k)
            n2 = int(len2 / k)

            if (n1 * str1[:k] == str1) and (n2 * str1[:k] == str2): # if the gcd string fully composes the 2 strings, it is valid gcd
                return True
            return False


        
        for i in range(min(len1, len2), 0, -1):
            if isgcd(i):
                return str1[:i]
        return ""

        # if (str1 + str2) != (str2 + str1): # then no common patterns
        #     return ""
        
        # lengcd = math.gcd(n, m)
        # gcd = str1[:lengcd]
        # return gcd


# 1431. Kids With the Greatest Number of Candies

# There are n kids with candies. You are given an integer array candies, where each candies[i] represents 
# the number of candies the ith kid has, and an integer extraCandies, denoting the number of extra candies that you have.
# Return a boolean array result of length n, where result[i] is true if, after giving the ith kid all the extraCandies, 
# they will have the greatest number of candies among all the kids, or false otherwise. Note that multiple kids can have 
# the greatest number of candies.

class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        n = len(candies)
        result = []
        initial_greatest = max(candies)

        for i in range(n):
            if (candies[i] + extraCandies )>= initial_greatest:
                result.append(True)
            else:
                result.append(False)
        return result



# 605. Can Place Flowers

# You have a long flowerbed in which some of the plots are planted, and some are not. However, flowers cannot be planted in adjacent plots.
# Given an integer array flowerbed containing 0's and 1's, where 0 means empty and 1 means not empty, and an integer n, return true if n 
# new flowers can be planted in the flowerbed without violating the no-adjacent-flowers rule and false otherwise.

class Solution:
    def canPlaceFlowers(self, flowerbed: List[int], n: int) -> bool:
        plantable = 0
        for i in range(len(flowerbed)):
            if flowerbed[i] == 0:
                if (i == 0) or (flowerbed[i-1] == 0):
                    if i == len(flowerbed)-1 or (flowerbed[i+1] == 0):
                        flowerbed[i] = 1
                        plantable += 1
        
        if n <= plantable:
            return True
        else:
            return False
        


# 345. Reverse Vowels of a String

# Given a string s, reverse only all the vowels in the string and return it.
# The vowels are 'a', 'e', 'i', 'o', and 'u', and they can appear in both lower and upper cases, more than once.

class Solution:
    def reverseVowels(self, s: str) -> str:
        vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
        indices = []
        for i, char in enumerate(s):
            if char in vowels:
                indices.append(i)

        string = list(s)
        left = 0
        right = len(indices) - 1

        while left < right:
            string[indices[left]], string[indices[right]] = string[indices[right]],string[indices[left]]
            left += 1
            right -= 1

        new_s = "".join(string)
        return new_s



# 151. Reverse Words in a String

# Given an input string s, reverse the order of the words.
# A word is defined as a sequence of non-space characters. The words in s will be separated by at least one space.
# Return a string of the words in reverse order concatenated by a single space.
# Note that s may contain leading or trailing spaces or multiple spaces between two words. The returned string should 
# only have a single space separating the words. Do not include any extra spaces.

class Solution:
    def reverseWords(self, s: str) -> str:
        s_list = s.split()
        s_list.reverse()
        rs_list = s_list
        reverse_str = " ".join(rs_list)
        return reverse_str
    


# 238. Product of Array Except Self

# Given an integer array nums, return an array answer such that answer[i] is equal to the product of all the elements of nums except nums[i].
# The product of any prefix or suffix of nums is guaranteed to fit in a 32-bit integer.
# You must write an algorithm that runs in O(n) time and without using the division operation.

from math import prod

class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        n = len(nums)
        answer = [1] * n
        prefix = 1
        suffix = 1
        for i in range(n):
            answer[i] *= prefix
            prefix *= nums[i]
        for j in range(n-1, -1, -1):
            answer[j] *= suffix
            suffix *= nums[j]
        
        return answer
```

```python
# 283. Move Zeroes

# Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.
# Note that you must do this in-place without making a copy of the array.

class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        lastNonZeroFoundBefore = 0
        for i in range(len(nums)):
            if nums[i] != 0: # if element is nonzero
                nums[lastNonZeroFoundBefore] = nums[i] # update lastfound location with nonzero element
                lastNonZeroFoundBefore += 1 # index lastfound forward
        for i in range(lastNonZeroFoundBefore, len(nums)): # fill remainder of list with 0s
            nums[i] = 0
        return nums
    



# 392. Is Subsequence

# Given two strings s and t, return true if s is a subsequence of t, or false otherwise.
# A subsequence of a string is a new string that is formed from the original string by deleting some (can be none) of the characters without 
# disturbing the relative positions of the remaining characters. (i.e., "ace" is a subsequence of "abcde" while "aec" is not).

class Solution:
    def isSubsequence(self, s: str, t: str) -> bool:
        p = 0
        for t_char in t:
            if p < len(s) and s[p] == t_char:
                p += 1
            else:
                continue
        if p == len(s):
            return True
        else:
            return False

# basically need to go through substring s, and check off one by one when they are found in order in t
```

### Sliding Window

```python
# 643. Maximum Average Subarray I

# You are given an integer array nums consisting of n elements, and an integer k.
# Find a contiguous subarray whose length is equal to k that has the maximum average value and return this value. 
# Any answer with a calculation error less than 10-5 will be accepted.

from statistics import mean
class Solution:
    def findMaxAverage(self, nums: List[int], k: int) -> float:
        n = len(nums)
        windowsum = sum(nums[:k])
        maxsum = windowsum
        for w in range(k, n):
            windowsum -= nums[w-k]
            windowsum += nums[w]
            if windowsum >= maxsum:
                maxsum = windowsum

        return maxsum / k
```

```python
# 1732. Find the Highest Altitude

# There is a biker going on a road trip. The road trip consists of n + 1 points at different altitudes. 
# The biker starts his trip on point 0 with altitude equal 0.
# You are given an integer array gain of length n where gain[i] is the net gain in altitude between points 
# i ​​​​and i + 1 for all (0 <= i < n). Return the highest altitude of a point.


class Solution:
    def largestAltitude(self, gain: List[int]) -> int:
        n = len(gain)
        altitudes = [0] * (n+1)
        for i in range(n):
            altitudes[i+1] = altitudes[i] + gain[i]
        return max(altitudes)






#724. Find Pivot Index

# Given an array of integers nums, calculate the pivot index of this array.
# The pivot index is the index where the sum of all the numbers strictly to the left of the index is equal to the sum of all the numbers strictly to the index's right.
# If the index is on the left edge of the array, then the left sum is 0 because there are no elements to the left. This also applies to the right edge of the array.
# Return the leftmost pivot index. If no such index exists, return -1.

class Solution:
    def pivotIndex(self, nums: List[int]) -> int:
        n = len(nums)
        for i in range(n):
            if i == 0:
                if sum(nums[1:]) == 0:
                    return i
            if i == n-1:
                if sum(nums[:n-1]) == 0:
                    return i
                else:
                    return -1
            if sum(nums[:i]) == sum(nums[i+1:]):
                return i
```

```python
### 933. Number of Recent Calls ###

# Implement the RecentCounter class which initializes the counter with zero recent requests.
# - RecentCounter() initiializes the counter with 0 recent requests
# - ping(t) Adds a new request at time t, where t represents some time in milliseconds, and returns the number
#   of requests that has happened in the past 3000 milliseconds (including the new request). Specifically, 
#   return the number of requests that have happened in the inclusive range [t - 3000, t].

from collections import deque

class RecentCounter(object):

    def __init__(self):
        self.queue = deque()

    def ping(self, t):
        """
        :type t: int
        :rtype: int
        """
        self.queue.append(t)
        while self.queue[0] < t -3000:
            self.queue.popleft()
        
        return len(self.queue)


obj = RecentCounter()
print(obj.ping(1))
print(obj.ping(2))
print(obj.ping(3))
print(obj.ping(3001))
print(obj.ping(3002))
```

```python
### 206. Reverse Linked List ###

# Given the head of a singly linked list, reverse the list, and return the reversed list.
# Modified to include full implementation of a Singly LinkedList Class.

# Definition for singly-linked list.


class LinkedList():
    class ListNode():
        def __init__(self, val=0, next=None):
            self.val = val
            self.next = next

    def __init__(self, values):
        self.head = self.ListNode(values[0])
        curr = self.head

        for i in range(len(values)-1):
            prev = curr
            curr = self.ListNode(values[i+1])
            prev.next = curr

    def reverseList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        curr = head
        prev = None

        while curr:
            next = curr.next
            curr.next = prev
            prev = curr
            curr = next
        return prev

    def printList(self, head):
        curr = head
        list_vals = []
        while curr:
            list_vals.append(str(curr.val))
            next = curr.next
            curr = next
        print(" -> ".join(list_vals))


list = [1, 2, 3, 4, 5]
linkedlist = LinkedList(list)
linkedlist.printList(linkedlist.head)

reversedhead = linkedlist.reverseList(linkedlist.head)
linkedlist.printList(reversedhead)
```

```python
### 104. Maximum Depth of Binary Tree ###

# Given the root of a binary tree, return its maximum depth.
# Recursion Help
#  - Base Case: "Whats the simplest version of the problem that can be answered?"
#      A: an empty tree --> maxdepth=0
#  - Recursive Case:"Can I solve this problem using smaller versions of itself?"
#      A: yes, if we know the max depth of the left subtree and right subtree we can just add 1 to that.


# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
from collections import deque

class Solution(object):
    def maxDepth(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """
        if not root:
            return 0
        return 1 + max(self.maxDepth(root.left), self.maxDepth(root.right))





### 872. Leaf-Similar Trees

# return true if the two trees have same leaf node sequences

# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def leafSimilar(self, root1, root2):
        """
        :type root1: Optional[TreeNode]
        :type root2: Optional[TreeNode]
        :rtype: bool
        """
        leafnodes1 = []
        leafnodes2 = []

        def dfs(root, leafnodes):
            if not root:
                return
            if not root.left and not root.right:
                leafnodes.append(root.val)

            dfs(root.left, leafnodes)
            dfs(root.right, leafnodes)
            
        dfs(root1, leafnodes1)
        dfs(root2, leafnodes2)

        return leafnodes1 == leafnodes2
```

```python
### 199. Binary Tree Right Side View ###

# Given root of Binary tree, assume you are standing on the right side of it. Return the values of the nodes you can see from top to bottom

from collections import deque
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def rightSideView(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: List[int]
        """
        if not root:
            return []
        
        queue = deque()
        queue.append(root)
        values = []

        while queue:
            levelsize = len(queue) # append the next level of tree to queue in current loop of for
            for i in range(levelsize):
                node = queue.popleft()
                
                if i == levelsize - 1: # if its rightmost in current level
                    values.append(node.val)

                if node.left: # add left children first so the last node in each level is the most right one
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
        return values





### 1161. Maximum Level Sum of a Binary Tree ###

# Given the root of a binary tree, the level of its root is 1, the level of its children is 2, and so on.
# Return the smallest level x such that the sum of all the values of nodes at level x is maximal.


from collections import deque
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def maxLevelSum(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """
        queue = deque()
        levels = {}

        queue.append(root)
        level = 0

        while queue:
            level += 1
            levelsum = 0
            for i in range(len(queue)):
                node = queue.popleft()
                levelsum += node.val
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            levels[level] = levelsum
        
        max_v = -9999
        max_k = -9999
        for k, v in levels.items():
            if v > max_v:
                max_v = v
                max_k = k
            elif v == max_v:
                if k < max_k:
                    max_v = v
                    max_k = k
        return max_k
        
```

```python
# 700. Search in a Binary Search Tree

# You are given the root of a binary search tree (BST) and an integer val.
# Find the node in the BST that the node's value equals val and return the subtree rooted with that node. If such a node does not exist, return null.

class Solution:
    def searchBST(self, root: Optional[TreeNode], val: int) -> Optional[TreeNode]:
        curr = root
        if curr.val == val:
            return curr
        elif curr.left and val < curr.val:
            curr = curr.left
            return self.searchBST(curr, val)
        elif curr.right and val > curr.val:
            curr = curr.right
            return self.searchBST(curr, val)
```

```python
### 1926. Nearest Exit from Entrance in Maze ###  BFS

# You are given an m x n matrix maze (0-indexed) with empty cells (represented as '.') and walls (represented as '+'). 
# You are also given the entrance of the maze, where entrance = [entrancerow, entrancecol] denotes the row and column 
# of the cell you are initially standing at.

# In one step, you can move one cell up, down, left, or right. You cannot step into a cell with a wall, and you cannot 
# step outside the maze. Your goal is to find the nearest exit from the entrance. An exit is defined as an empty cell 
# that is at the border of the maze. The entrance does not count as an exit.

# Return the number of steps in the shortest path from the entrance to the nearest exit, or -1 if no such path exists.

from collections import deque

class Solution(object):
    def nearestExit(self, maze, entrance):
        """
        :type maze: List[List[str]]
        :type entrance: List[int]
        :rtype: int
        """
        directions = [(0, 1), (1, 0), (-1, 0), (0, -1)]
        queue = deque()
        steps = 0
        visited = set()
        start = tuple(entrance)
        tot_rows = len(maze)
        tot_cols = len(maze[0])

        queue.append((start, steps))
        visited.add(start)

        while queue:
            [(cr, cc), steps] = queue.popleft()

            for d in directions:
                nr, nc = cr + d[0], cc + d[1]
                if (0 <= nr < tot_rows) and (0 <= nc < tot_cols): # if valid cell in maze
                    if (maze[nr][nc] == '.') and ((nr, nc) not in visited): # if empty and not visited before
                        if ((nr == 0) or (nr == tot_rows - 1) or (nc == 0) or (nc == tot_cols - 1)) and ((nr, nc) != start): # if border cell space and not entrance
                            return steps + 1
                        queue.append(((nr, nc), steps+1))
                        visited.add((nr, nc))
        return -1




### 994. Rotting Oranges

# BFS Search
# 0 - empty cell
# 1 - fresh orange
# 2 - rotten orange

# Every minute, any fresh orange that is 4-directionally adjacent to a rotten orange becomes rotten.
# Return the minimum number of minutes that must elapse until no cell has a fresh orange. If this is impossible, return -1.

from collections import deque


def orangesRotting(grid):
    """
    :type grid: List[List[int]]
    :rtype: int
    """
    h = len(grid)
    w = len(grid[0])
    directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]

    queue = deque()
    minutes = 0

    fresh = 0
    for r in range(h):
        for c in range(w):
            if grid[r][c] == 2:
                queue.append((r, c))
            elif grid[r][c] == 1:
                fresh += 1

    while queue and fresh != 0:
        for i in range(len(queue)): # all the rotten orange locations in the current minute
            (cr, cc) = queue.popleft()

            for (dr, dc) in directions:
                (nr, nc) = (cr + dr, cc + dc)
                if (0 <= nr < h) and (0 <= nc < w): # if valid cell
                    if grid[nr][nc] == 1: # if fresh turn to rotten since neighbor rotten
                        grid[nr][nc] = 2
                        queue.append((nr, nc))
                        fresh -= 1
        minutes += 1
        
    if fresh == 0:
        return minutes
    else:
        return -1


grid = [[2,1,1],[1,1,0],[0,1,1]]
print(orangesRotting(grid))
```

```python
### 374. Guess Number Higher or Lower ###

# I pick a number from 1 to n. You have to guess which number I picked.
# Every time you guess wrong, I will tell you whether the number I picked is higher or lower than your guess.
# You call a pre-defined API int guess(int num), which returns three possible results:

# -1: Your guess is higher than the number I picked (i.e. num > pick).
# 1: Your guess is lower than the number I picked (i.e. num < pick).
# 0: your guess is equal to the number I picked (i.e. num == pick).
# Return the number that I picked.

# The guess API is already defined for you.
# @param num, your guess
# @return -1 if num is higher than the picked number
#          1 if num is lower than the picked number
#          otherwise return 0
# def guess(num):

class Solution(object):
    def guessNumber(self, n):
        """
        :type n: int
        :rtype: int
        """
        left = 1
        right = n
        mid = (left + right) // 2

        while guess(mid) != 0:
            mid = (left + right) // 2
            if guess(mid) == -1: # if guess is higher
                left = left
                right = mid - 1 # -1 bcuz u shud exclude prev mid guess, u know its lower than prev mid
            elif guess(mid) == 1: # if guess is lower
                left = mid + 1 # +1 bcuz u shud exclude prev mid guess, u know its higher than prev mid
                right = right
        return mid
    

    
### 2300. Successful Pairs of Spells and Potions ###

# You are given two positive integer arrays spells and potions, of length n and m respectively, 
# where spells[i] represents the strength of the ith spell and potions[j] represents the strength of the jth potion.

# You are also given an integer success. A spell and potion pair is considered successful if the product of their strengths is at least success.
# Return an integer array pairs of length n where pairs[i] is the number of potions that will form a successful pair with the ith spell.

class Solution(object):
    def successfulPairs(self, spells, potions, success):
        """
        :type spells: List[int]
        :type potions: List[int]
        :type success: int
        :rtype: List[int]
        """
        n = len(spells)
        m = len(potions)
        potions.sort()
        pairs = [0] * n

        for i in range(n): #spells
            
            left = 0
            right = m-1

            while left <= right: #answer could be anywhere between left and right inclusive
                mid = (left + right) // 2
                if (spells[i] * potions[mid]) < success: #not successful if product is less than success, shift search area to right
                    left = mid + 1
                    right = right
                    
                elif (spells[i] * potions[mid]) >= success: #successful if product is equal or greater than success, 
                # shift search area to left to get the smallest mid possible
                    left = left
                    right = mid - 1
                    
            #while loop ends when left > right, left will be pointing at the first valid index
            success_count = m - left
            pairs[i] = success_count
        
        return pairs
    

### 162. Find Peak Element ###

# A peak element is an element that is strictly greater than its neighbors.
# Given a 0-indexed integer array nums, find a peak element, and return its index. If the array contains multiple peaks, return the index to any of the peaks.
# You may imagine that nums[-1] = nums[n] = -∞. In other words, an element is always considered to be strictly greater than a neighbor that is outside the array.
# You must write an algorithm that runs in O(log n) time.

class Solution(object):
    def findPeakElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n = len(nums)
        left = 0
        right = n-1

        while left <= right:
            mid = (left + right) // 2
            if ((mid - 1) < 0) and ((mid + 1) > (n - 1)): # Edge case of 1 element list [1]
                return mid
            elif (mid - 1) < 0: # Edge case of left neighbor being outside array
                if nums[mid + 1] < nums[mid]:
                    return mid
                else:
                    left = mid + 1
                    continue
            elif (mid + 1) > (n-1): # Edge case of right neighbor being outside array
                if nums[mid-1] < nums[mid]:
                    return mid
                else:
                    right = mid - 1
                    continue
            if nums[mid - 1] > nums[mid]: # if left neighbor greater, search left half
                right = mid - 1
            elif nums[mid + 1] > nums[mid]: # if right neighbor greater, search right half
                left = mid + 1
            else: # if neither greater you found the peak
                return mid


# MUCH SHORTER SOLUTION - INSTEAD OF EXPLICIT EDGE CASE HANDLING, JUST SUBSTITUTE LEFT AND RIGHT VALUE WITH -INFINITY
class Solution(object):
    def findPeakElement(self, nums):
        left, right = 0, len(nums) - 1

        while left <= right:
            mid = (left + right) // 2
            left_val = float('-inf') if mid == 0 else nums[mid - 1]
            right_val = float('-inf') if mid == len(nums) - 1 else nums[mid + 1]

            if nums[mid] > left_val and nums[mid] > right_val:
                return mid
            elif nums[mid] < right_val:
                left = mid + 1
            else:
                right = mid - 1
```

```python
# 338. Counting Bits

# Given an integer n, return an array ans of length n + 1 such that for each i (0 <= i <= n), ans[i] is the number of 1's in the binary representation of i.

class Solution:
    def countBits(self, n: int) -> List[int]:
        ans = [0] * (n+1)
        for i in range(n+1):
            ans[i] = bin(i).count('1')
        return ans
    


# 136. Single Number

# Given a non-empty array of integers nums, every element appears twice except for one. Find that single one.
# You must implement a solution with a linear runtime complexity and use only constant extra space.

from collections import Counter
class Solution:
    def singleNumber(self, nums: List[int]) -> int:
        unicount = Counter(nums)
        
        for n in nums:
            if unicount[n] == 1:
                return n
```

```python
# A robot operates in a 2D warehouse grid, represented as a matrix or a list of lists. 

# 0 : "empty space"
# 1 : "item to pick up"
# -1: "obstacle"

# The robot starts at (0, 0), write a function that returns the shortest path as a list of (row, column) tuples.

# We need to visit ever cell and keep track of the ones we visited already. We also need to track the shortest path to get to each one. Use a queue to implement a BFS algorithm.
# Check the nearest neighbors of the current cell, if they are not visited and not an obstacle, add them to the queue. If we find a cell with a 1, return the path taken to get there.
# The queue makes sure we check the closest neighbors first, so we will always find the shortest path to the item.
from collections import deque


def pickup_plan(grid):
    '''
    Use a BFS Algorithm to find the shortest path.
    '''
    neighbors = [(0, 1), (1, 0), (0, -1), (-1, 0)]

    start = (0, 0)
    visited = set()
    visited.add(start)

    queue = deque()
    queue.append((start, [start]))  # (current position, path taken)

    while queue:
        (r, c), path = queue.popleft()

        for (dr, dc) in neighbors:
            nr, nc = r + dr, c + dc
            if 0 <= nr < len(grid) and 0 <= nc < len(grid[0]):  #if the new position is in the warehouse
                if (nr, nc) not in visited and grid[nr][nc] != -1:
                    if grid[nr][nc] == 1:
                        return path + [(nr, nc)]
                    visited.add((nr, nc))
                    queue.append(((nr, nc), path + [(nr, nc)]))


grid = [
    [0, 0, 0, 0],
    [0, -1, -1, 0],
    [0, 1, 0, 0],
]

print(pickup_plan(grid))
```
